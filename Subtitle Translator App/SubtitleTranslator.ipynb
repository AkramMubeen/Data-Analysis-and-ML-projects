```python
import datetime

def to_srt_time(time):
    """
    Convert time in seconds to SRT time format (hours:minutes:seconds,milliseconds).
    """
    milliseconds = int((time - int(time)) * 1000)
    s, ms = divmod(milliseconds, 1000)
    m, s = divmod(int(time), 60)
    h, m = divmod(m, 60)
    return "{:02d}:{:02d}:{:02d},{:03d}".format(h, m, s, ms)

with open("check.srt", "w") as f:
    for i, segment in enumerate(result_aligned["segments"], start=1):
        start = to_srt_time(segment["start"])
        end = to_srt_time(segment["end"])
        text = segment["text"]
        f.write(f"{i}\n{start} --> {end}\n{text}\n\n")
```

```python
import os

input_file_path = '/content/check.srt'
output_file_path = 'output/'

# Create the output directory if it doesn't exist
os.makedirs(output_file_path, exist_ok=True)

with open(input_file_path, 'r') as input_file:
    lines = input_file.readlines()

# Count the number of lines
num_lines = len(lines)
print(f"Number of lines in the input file: {num_lines}")
```

```python
import math
split_size = 1000  # Number of lines per part
num_parts = math.ceil(len(lines)/split_size)  # Calculate the number of parts
chunks = []  # Create an empty list to store output file paths

output_directory = 'output'  # Change this to your desired output directory
os.makedirs(output_directory, exist_ok=True)

for part_num in range(num_parts):
    start_index = part_num * split_size
    end_index = (part_num + 1) * split_size

    part_lines = lines[start_index:end_index]

    output_file_name = f'part{part_num + 1}.txt'
    output_file_path = os.path.join(output_directory, output_file_name)

    with open(output_file_path, 'w') as output_file:
        output_file.writelines(part_lines)
    chunks.append(output_file_path)  # Append the output file path to the list

print("Chunks:", chunks)
```

```python
import gradio as gr
import openai
import whisper
import whisperx
import tiktoken
import datetime
import math
import os

# Set your OpenAI API key
openai.api_key = "OPENAI KEY"

# Function to count tokens in input text
def count_tokens(input_text, model_name="gpt-3.5-turbo-16k"):
    enc = tiktoken.encoding_for_model(model_name)
    return len(enc.encode(input_text))

# Function to open a file and return its content
def open_file(filepath):
    with open(filepath, 'r', encoding='utf-8') as infile:
        return infile.read()

# Function to save content to a file
def save_file(content, filepath):
    with open(filepath, 'w', encoding='utf-8') as outfile:
        outfile.write(content)

# Function to transcribe audio
def transcribe_audio(audio_file_path):
    print("Transcribing audio...")
    model = whisper.load_model("large-v2")
    result = model.transcribe(audio_file_path)
    print("Audio transcription completed.")
    return result["segments"]

# Function to align segments
def align_segments(segments, audio_file_path, device):
    print("Aligning segments...")
    model_a, metadata = whisperx.load_align_model(
        language_code=metadata["language"],
        device=device
    )
    result = whisperx.align(segments, model_a, metadata, audio_file_path, device)
    print("Segment alignment completed.")
    return result

# Function to convert aligned segments to SRT format
def convert_to_srt(segments):
    print("Converting to SRT format...")
    subtitles = []
    for i, segment in enumerate(segments, start=1):
        start = to_srt_time(segment["start"])
        end = to_srt_time(segment["end"])
        text = segment["text"]
        subtitle = f"{i}\n{start} --> {end}\n{text}\n\n"
        subtitles.append(subtitle)
    print("Conversion to SRT format completed.")
    return subtitles

# Function to convert time to SRT format
def to_srt_time(time):
    milliseconds = int((time - int(time)) * 1000)
    s, ms = divmod(milliseconds, 1000)
    m, s = divmod(int(time), 60)
    h, m = divmod(m, 60)
    return "{:02d}:{:02d}:{:02d},{:03d}".format(h, m, s, ms)

# Function to split subtitles into chunks
def split_subtitles(subtitles, max_tokens_per_chunk=15500):
    print("Splitting subtitles into chunks...")
    chunks = []
    current_chunk = []
    current_token_count = 0

    for subtitle in subtitles:
        subtitle_tokens = count_tokens(subtitle)

        if current_token_count + subtitle_tokens > max_tokens_per_chunk:
            chunks.append(current_chunk)
            current_chunk = []
            current_token_count = 0

        current_chunk.append(subtitle)
        current_token_count += subtitle_tokens

    if current_chunk:
        chunks.append(current_chunk)

    print("Subtitles split into chunks.")
    return chunks

# Function to generate OpenAI responses
def generate_openai_responses(subtitle_chunks, instruct_prompt, model_name="gpt-3.5-turbo-16k"):
    print("Generating OpenAI responses...")
    responses = []

    for count, chunk in enumerate(subtitle_chunks, start=1):
        chunk_text = open_file(chunk)
        request = instruct_prompt + chunk_text

        messages = [
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": request}
        ]

        chat_output = openai.ChatCompletion.create(model=model_name, messages=messages)

        response = chat_output.choices[0].message.content
        responses.append(response)

    print("OpenAI responses generated.")
    return responses

# Define Gradio interface
def audio_transcription(input_audio):
    # Create a temporary directory to store files
    temp_dir = "./temp"
    os.makedirs(temp_dir, exist_ok=True)

    # Save the uploaded audio file
    audio_file_path = os.path.join(temp_dir, "audio.wav")
    with open(audio_file_path, 'wb') as audio_file:
        audio_file.write(input_audio)

    # Transcribe the uploaded audio file
    transcription = transcribe_audio(audio_file_path)

    # Align the transcription segments
    aligned_transcription = align_segments(
        transcription,
        audio_file_path,
        device="cuda"
    )

    # Convert the aligned result to subtitle format
    subtitles = convert_to_srt(aligned_transcription)

    # Count tokens and lines
    token_count = count_tokens("\n".join(subtitles))
    num_lines = len(subtitles)

    # Check if token count exceeds the limit
    if token_count > 15500:
        print("Token count exceeds 15,500. Splitting subtitles into chunks...")

        instruct_prompt = "Translate the following subtitles into English. While translating, please try to recognize entities and places. Additionally, ensure there are no spelling mistakes, particularly when it comes to city and location names. Kindly review the translation carefully and provide your response."
        subtitle_chunks = split_subtitles(subtitles)
        openai_responses = generate_openai_responses(subtitle_chunks, instruct_prompt)

        # Save the OpenAI responses to a file
        responses_text = '\n\n'.join(openai_responses)
        save_file(responses_text, os.path.join(temp_dir, 'openai_responses.txt'))

    return transcription

# Create Gradio interface
gr.Interface(
    fn=audio_transcription,
    inputs=gr.inputs.Audio(label="Upload an audio file"),
    outputs=gr.outputs.Textbox(label="Transcription"),
    title="Audio Transcription and Subtitle Generation",
    live=True,
).launch()
```
