```python
address = '/content/gdrive/MyDrive/HousePrices/'
scaler_path = address + "scaler/scaler.pkl"

def load_and_apply_scaler(df,columns_to_scale,file_path):
    with open(file_path, 'rb') as file:
        scaler = pickle.load(file)
    df[columns_to_scale] = scaler.transform(df[columns_to_scale])
    return df
```

```python
df['cluster_id'] = df['cluster_id'].astype('int64')
```

```python
# Cluster 0
with open('/content/gdrive/MyDrive/HousePrices/models/XGBRegressor_0.pkl', 'rb') as file:
  model_0 = pickle.load(file)
# Cluster 1
with open('/content/gdrive/MyDrive/HousePrices/models/XGBRegressor_1.pkl', 'rb') as file:
  model_1 = pickle.load(file)
# Cluster 2
with open('/content/gdrive/MyDrive/HousePrices/models/XGBRegressor_2.pkl', 'rb') as file:
  model_2 = pickle.load(file)
# Cluster 3
with open('/content/gdrive/MyDrive/HousePrices/models/RandomForestRegressor_3.pkl', 'rb') as file:
  model_3 = pickle.load(file)
```

```python
# Load the model_predict function
def model_predict(cluster_id, features):
  cluster_id = cluster_id[0]
  if cluster_id == 0:
        # Cluster 0
      model = model_0
  elif cluster_id == 1:
        # Cluster 1
      model = model_1
  elif cluster_id == 2:
        # Cluster 2
      model = model_2
  elif cluster_id == 3:
        # Cluster 3
      model = model_3

    # Make predictions using the selected model
  prediction = model.predict(features)

  return prediction
```

```python
predictions_list = []
for i in range(len(df)):
  dfx = df.iloc[i]
  dfx = dfx.to_frame()
  dfx = dfx.T
  cluster_id = list(dfx['cluster_id'])
  dfx = dfx.drop('cluster_id',axis=1)
  prediction = model_predict(cluster_id, dfx)
  predictions_list.append(prediction[0])
```

```python
fd = pd.DataFrame({'Id': Id, 'SalePrice': predictions_list})
```

```python
fd.to_csv(address + "predictions.csv",index=False)
```
